```python
def get_score(model,X_train,X_test,Y_train,Y_test):
    model.fit(X_train,Y_train)
    return model.score(X_test,Y_test)
```

```python
dataset.sort_values(['Class'],ascending=True,inplace=True)
```

# Creating Sub Dataset

As the provided data set is highly imbalanced we make use of sampling technique, to create a new dataset which is balanced.
If we use the original datatset it leads to

> Overfitting: Due to overfitting our models will be more inclined to predict the ouput as non fraud, but our main aim is deal with the fraud detection.
> Wrong Correleation: By having an imbalance dataset we cannot estimate or figure out perfect correleation among our features

Hence, we create a sub dataset with equal proportion of each variety i.e we maintain a ratio of 1:1 or 50-50 and take 492 fraud cases and 492 random non fraud cases, to balance our new dataset, and we use this dataset to train our model

```python
fraud = dataset.loc[dataset['Class']==1]

non_fraud = dataset.iloc[0:492,:]
```

```python
new_data = pd.concat([fraud,non_fraud])
```

Separating dependent and Independent variable:

```python
x = new_data.iloc[:,0:30].values

y = new_data.iloc[:,30:].values
```

range from numbers as high as 17765 to as low as -0.1563.

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

# Model Selection

We now test diffrent algorithms to examine which one suits best to the manually fabricated dataset.

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='liblinear',multi_class='ovr')
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=10)
```

```python
from sklearn.svm import SVC

s = SVC(kernel='rbf',gamma='auto')
```

# Implementing Cross Validation Techniques

```python
from sklearn.model_selection import StratifiedKFold
fold = StratifiedKFold(n_splits=3)
```

```python
scores_log = []
scores_smv = []
scores_rf = []
```

```python
for train_index,test_index in fold.split(x,y):
    X_train,X_test,Y_train,Y_test = x[train_index],x[test_index],y[train_index],y[test_index]
    scores_log.append(get_score(lr,X_train,X_test,Y_train.ravel(),Y_test))
    scores_smv.append(get_score(s,X_train,X_test,Y_train.ravel(),Y_test))
    scores_rf.append(get_score(rf,X_train,X_test,Y_train.ravel(),Y_test))
```

AS RANDOM FOREST CLASSIFIER GIVES THE BEST ACCURACY THEN IT IS CHOSEN FOR MODEL BUILDING
HERE IS THE AREA UNDER PRECISION-RECALL GRAPH

```python
from sklearn.ensemble import RandomForestClassifier
dt=RandomForestClassifier(criterion='entropy',n_estimators=10,random_state=0,n_jobs=-1)
dt.fit(x_train,y_train)
```

```python
ypred=dt.predict(x_test)
```

```python
from sklearn.metrics import average_precision_score
average_precision = average_precision_score(y_test, ypred)

from sklearn.metrics import precision_recall_curve
from sklearn.metrics import plot_precision_recall_curve
import matplotlib.pyplot as plt

disp = plot_precision_recall_curve(dt, x_test, y_test)
disp.ax_.set_title('2-class Precision-Recall curve: '
                   'AP={0:0.2f}'.format(average_precision))
```
